# Hierarchical Clustering

```python
from sklearn.preprocessing import StandardScaler 
```

```python
sc = StandardScaler()
y = sc.fit_transform(x)
```

```python
y
```

```python
import scipy.cluster.hierarchy as sch
```

```python
lk = sch.linkage(y,method='complete')
dendro = sch.dendrogram(lk)
plt.show()
```

# K-Means

```python
from sklearn.cluster import KMeans
```

```python
km = KMeans()
```

```python
ab = km.fit_predict(x)
```

```python
km.cluster_centers_
```

```python
x
```

```python
plt.scatter(y[:,0],y[:,2],c=ab,cmap='viridis')
plt.show()
```

```python
wcss =  []
for i in range(1,11):
    km = KMeans(n_clusters=i)
    km.fit_predict(x)
    wcss.append(km.inertia_)
```

```python
plt.plot(range(1,11),wcss)
plt.show()
```

# DB-Scan

```python
df = pd.read_csv('https://raw.githubusercontent.com/yashvardhanpatil/Assignment-Dataset/main/Clustering/crime_data.csv')
df
```

```python
df.drop(columns='Unnamed: 0',inplace=True)
df
```

```python
x = sc.fit_transform(df)
x
```

```python
from sklearn.cluster import DBSCAN
```

```python
db  = DBSCAN(eps=1,min_samples=4) 
db.fit_predict(x)
```

```python
db.labels_
```

```python
df['DBscan_Cluster']= db.labels_
```

```python
df
```

```python
df.groupby('DBscan_Cluster').agg(['mean']).reset_index()
```

```python
plt.scatter(df['DBscan_Cluster'],df['UrbanPop'],c=db.labels_)
plt.show()
```
