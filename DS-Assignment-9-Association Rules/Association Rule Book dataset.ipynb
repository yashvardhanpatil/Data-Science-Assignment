```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(style='darkgrid',palette='rainbow')
from mlxtend.frequent_patterns import association_rules,apriori
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/yashvardhanpatil/Assignment-Dataset/main/Association%20Rules/book.csv')
df
```

```python
df.describe()
```

```python
df.isna().sum()
```

```python
df.info()
```

### with Support value (0.1)

```python
ap = apriori(df,min_support=0.1,use_colnames=True)
ap
```

```python
ass_rule = association_rules(ap,metric='lift')
ass_rule
```

```python
sns.scatterplot(x=ass_rule['support'],y=ass_rule['confidence'])
plt.title('Scatter Plot')
```

### With support-Value (0.05)

```python
ap_2 = apriori(df,min_support=0.05,use_colnames=True)
ap_2
```
